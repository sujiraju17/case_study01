# Case study on Presentation of data

# 1. Read the dataset to the python environment.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

c_data=pd.read_csv('cars_data.csv')
c_data
```

# 2. Check for the null values present in the dataset.

```python
c_data.isnull()
```

# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
plt.figure(figsize=(8,8))
sns.countplot(x='Buyer Gender',data=c_data)
plt.title('Male vs Female buyers',fontsize=18,color='red')
```

# 4. Find the top 5 cars based on their sales price.

```python
c_data['Sale Price'].nlargest(n=5)
```

# 5. Find the least 5 cars based on their Resell price.

```python
c_data['Resell Price'].nsmallest(n=5)
```
